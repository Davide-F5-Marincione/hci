```python
import pandas as pd

df = pd.read_csv("namelist.csv")

df.head()
```

```python
a = "pn_generic_planets sn_corvette	sn_destroyer	sn_frigate	sn_cruiser	sn_battle_cruiser	sn_striker_cruiser	sn_battleship	sn_carrier	sn_dreadnought	sn_titan	sn_flagship	sn_juggernaut	sn_colossus	sn_colonizer	sn_sponsored_colonizer	sn_constructor	sn_exploration_ship	sn_science"
b = a.split()
print(b)
```

```python
names = set()
AO = set("qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM")
for col in b:
    curr_names = df[col].tolist()
    for name in curr_names:
        if type(name) is not str:
            continue
        if set(name).issubset(AO):
            names.add(name.lower())
```

```python
names.update({"rubicante", "golbez", "biggs", "wedge", "ravana", "morgoth", "finrod", "aznable", "unicorn", "zaku"})
print(names)
```

```python
import pickle
print(len(names))
with open("names.pickle", "+wb") as file:
    pickle.dump(list(names), file)
```
